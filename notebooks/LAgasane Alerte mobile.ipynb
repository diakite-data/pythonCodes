```python
def hlr_bscs_in1(folder_alerte) :
    path = f"/Users/youssouf/Downloads/{folder_alerte}/"
    list_path = list(paths.list_files(path))
    print(list_path)
    L_BSCS = []
    L_IN = []
    L_HLR = []
    L_NAME = []
    df_hlr = None
    df_bscs = None
    df_in = None
    n = None 
    
    for path in list_path :
        for i in ["HLR", "BSCS", "IN"] :
            try :
                df = pd.read_excel(path, skiprows = 4, sheet_name=f"Détail {i}")      
                if (i == "HLR") :
                    L_HLR.append(df)
                elif (i == "BSCS") :
                    L_BSCS.append(df)
                else :
                    L_IN.append(df)
            except :
                print(f"path doesn't exist {path} - {i}")
        n = path.split("Alerte_detail_rappo_Mobile_")[1].split("_17-11-2021.xlsx")[0].split("_")[1] 
        
        
        L_NAME.append(n)
    
    #L_BSCS1 = L_BSCS
    #L_HLR1 = L_HLR
    #L_IN1 = L_IN
    
    try : 
        df_bscs = pd.concat(L_BSCS)
        df_bscs["Alerte"] = df_bscs.shape[0]*[L_NAME[0]]
    except :
        print("list vide BSCS")
     
    try :
        df_hlr = pd.concat(L_HLR)
        df_hlr["Alerte"] = df_hlr.shape[0]*[L_NAME[0]]
    except :
        print("list vide HLR")
     
    try :
        df_in = pd.concat(L_IN)
        df_in["Alerte"] = df_in.shape[0]*[L_NAME[0]]
    except :
        print("list vide IN")
        
    return df_bscs, df_hlr, df_in 



```

```python
def f1(T) :
    for i in range(1,len(T)) :
        x = T[i]
        j = 0
        while(j <= (i-1) and T[j] < x) :
            j = j + 1
        for k in range(i-1,j,-1) : 
            T[k+1] = T[k]
        T[j] = x
    return T        
```

```python
f1([4,7,2,1])
```

```python
def f2(T) :
    for i in range(1,len(T)) : 
        x = T[i]
        j = i - 1
        while(j>=0 and T[j]>x) :
            T[j+1] = T[j]
            j = j-1
        T[j+1] = x
    return T
```

```python
f2([4,7,2,1])
```

```python
f2([98, 22, 15, 32, 2, 74, 63, 70])
```

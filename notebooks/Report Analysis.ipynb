## Outliers

Pour choisir les outliers, nous avons sélectionné des variables.
  - nombre d'appels
  - nombre d'appels nuits 
  - nombre d'appels internationals
  - nombre sms

Car ces variables sont corrélées avec la plupart des autres variables.

Pour déterminer les outliers la règle suivante a été suivante a été appliquée pour chaque variable :
 - s = df['var']
 - q1 = s.quantile(0.25)
 - q3 = s.quantile(0.75)
 - iqr = q3 - q1
 - iqr_lower = q1 - 1.5 * iqr
 - iqr_upper = q3 + 1.5 * iqr
 - outliers = dict(s[(s < iqr_lower) | (s > iqr_upper)])

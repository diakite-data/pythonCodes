```python
def forte_corr(seuil = 0.5 , df = df_ , method = 'pearson') :
    #list_plot = df.columns[2:]
    list_plot = df.columns
    L_index = []
    List_var = []
    List_corr = []
    
    cor = df[list_plot].corr(method = method)
    cor = cor[cor > seuil]
    index = cor.index
    for col in cor.columns : 
        for indice , row in enumerate(cor[col]):
            if (not np.isnan(row)) and (row != 1 ) :
                L_index.append(index[indice])
                List_var.append(col)
                List_corr.append(row)
    #[(x , y , z) for x , y , z in zip (L_index , List_var ,List_corr )]                                                    
    df = pd.DataFrame({'Var1':L_index , 'Var2':List_var , 'Correlation':List_corr}).sort_values('Correlation').reset_index()
    df = df.drop('index' , 1)
    
    return df[[True if x%2 == 0 else False for x in range(df.shape[0])]]
```

```python
df_corr = forte_corr()
```

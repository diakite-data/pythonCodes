```python
def somme_rec(l) : 
    
    flatten_list = []
    for i in l :
        if type(i) == int :
            flatten_list.append(i)
        elif type(i) == list :
            for j in i :
                if type(j) == int :
                    flatten_list.append(j)
                elif type(j) == list :
                    for k in j : 
                        flatten_list.append(k)
                        
    return sum(flatten_list)
```

```python
def drop_duplicates(df, columns=None):
    """Return a :class:`DataFrame` object with no duplicates in the given columns.
    .. warning:: The resulting dataframe will be in memory, use with caution.
    :param columns: Column or list of column to remove duplicates by, default to all columns.
    :return: :class:`DataFrame` object with duplicates filtered away.
    """
    if columns is None:
        columns = df.get_column_names()
    if type(columns) is str:
        columns = [columns]
    return df.groupby(columns, agg={'__hidden_count': vaex.agg.count()}).drop('__hidden_count')
```

```python
import pandas as pd

def aggregate_times(df):
    # Calculer le temps moyen et le temps mdian pour chaque sprinter
    mean_times = df.groupby(['FirstNan', 'LastName'])['Time'].mean()
    median_times = df.groupby(['FirstNan', 'LastName'])['Time'].median()
    
    # Crer un nouveau DataFrame avec les colonnes FullName, MeanTime et MedianTime
    new_df = pd.DataFrame({'FullName': [f"{first} {last}" for first, last in mean_times.index],
                           'MeanTime': mean_times.values,
                           'MedianTime': median_times.values})
    
    # Trier le DataFrame par ordre alphabtique de FullName
    new_df = new_df.sort_values('FullName')
    
    return new_df
```

```python
import numpy as np

def cluster_points(points, cluster_centers):
    clusters = {center_name: [] for center_name in cluster_centers}
    
    for point_idx, point in enumerate(points):
        closest_center = None
        closest_distance = np.inf
        
        for center_name, center_coords in cluster_centers.items():
            distance = np.linalg.norm(point - center_coords)
            
            if distance < closest_distance:
                closest_center = center_name
                closest_distance = distance
        
        clusters[closest_center].append(point_idx)
    
    return clusters

```

```python
def compute_recall(m):
    recall = m.true_positive / (m.true_positive + m.false_negative)
    return recall

def compute_precision(m):
    precision = m.true_positive / (m.true_positive + m.false_positive)
    return precision

def compute_specificity(m):
    specificity = m.true_negative / (m.true_negative + m.false_positive)
    return specificity

def compute_accuracy(m):
    accuracy = (m.true_positive + m.true_negative) / (m.true_positive + m.true_negative + m.false_positive + m.false_negative)
    return accuracy
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Example lists
list1 = ["2023-06-27", "2023-06-28", "2023-06-29"]
list2 = ['09:13:00', '10:15:00', '08:45:00']

# Create a DataFrame
df = pd.DataFrame({'day': list1, 'hour': list2})

# Convert 'day' column to datetime format
df['day'] = pd.to_datetime(df['day'])

# Convert 'hour' column to datetime format
df['hour'] = pd.to_datetime(df['hour'], format='%H:%M:%S').dt.time

# Set the y-axis interval
start_time = datetime.strptime("06:00:00", "%H:%M:%S").time()
end_time = datetime.strptime("23:59:00", "%H:%M:%S").time()
interval = timedelta(hours=1)

# Generate datetime values for the y-axis
y_values = []
current_time = datetime(2023, 1, 1, start_time.hour, start_time.minute, start_time.second)
while current_time.time() <= end_time:
    y_values.append(current_time)
    current_time += interval

# Plot the data
plt.plot(df['day'], df['hour'], marker='o')

# Set the x-axis label
plt.xlabel('Day')

# Set the y-axis label
plt.ylabel('Hour')

# Set the y-axis interval ticks and labels
plt.yticks(y_values)

# Rotate x-axis tick labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
from datetime import datetime

def cast_to_date(date_str):
    # Convert date string to date object
    date_object = datetime.strptime(date_str, "%d-%m-%Y").date()
    return date_object

def cast_to_date2(date_str):
    
    date_str = date_str.split(" ")[0]
    
    if "/" in date_str :
        date_object = datetime.strptime(date_str, "%d/%m/%Y").date()
        return date_object
    
    elif "-" in date_str and len(str(date_str.split("-")[-1])) == 4 :
        return datetime.strptime(date_str, "%d-%m-%Y").date()
    
    return datetime.strptime(date_str, "%Y-%m-%d").date()

def cast_to_date3(date_str):
    # Convert date string to date object
    date_object = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S").date()
    return date_object
```

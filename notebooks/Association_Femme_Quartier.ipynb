```python
import pandas as pd
import numpy as np
```

```python
def femme_quartier(path) :
    Nom = []
    Prenoms = []
    DN = []
    CON = []
    PR = []
    PI = []
    NP = []
    with open(path, 'r') as in_file:
        
            lines = in_file.readlines()
    
    for line in lines:
        if "NOM :" in line :
            Nom.append(line.replace("NOM :", ""))
        
        elif "PRENOMS :" in line :
            Prenoms.append(line.replace("PRENOMS :", ""))
            
        elif "PROFESSION :" in line :
            PR.append(line.replace("PROFESSION :", ""))
            
        elif "DATE ET LIEU DE NAISSANCE :" in line :
            DN.append(line.replace("DATE ET LIEU DE NAISSANCE :", ""))
               
        elif "CONTACTS :" in line :
            CON.append(line.replace("CONTACTS :", ""))
        
        elif "PIÈCE D'IDENTIFICATION :" in line :
            PI.append(line.replace("PIÈCE D'IDENTIFICATION :", ""))

        elif "NUMERO DE LA PIÈCE" in line :
            NP.append(line.replace("NUMERO DE LA PIÈCE", ""))

    print([len(Nom), len(Prenoms), len(PR), len(DN), len(CON), len(PI), len(NP)])
    #print(Sexe)
    #print("Nombre de femme " + str(Sexe.count("\t F\n")))
    #print("Nombre de homme " + str(Sexe.count("\t M\n")))
    
    df = pd.DataFrame({'NOM': Nom , 'PRENOMS': Prenoms , "PROFESSION": PR,  'DATE_ET_LIEU_NAISSANCE': DN 
                   
                   ,'CONTACTS': CON,'PIECE_IDENTIFICATION': PI,'NUMERO_PIECE': NP})
      
    df["NOM"] = df["NOM"].str.replace('\t' ,'', regex=True)
    df["NOM"] = df["NOM"].str.replace('\n' ,'', regex=True)

    df["PRENOMS"] = df["PRENOMS"].str.replace('\t' ,'', regex=True)
    df["PRENOMS"] = df["PRENOMS"].str.replace('\n' ,'', regex=True)
    
    df["PROFESSION"] = df["PROFESSION"].str.replace('\t' ,'', regex=True)
    df["PROFESSION"] = df["PROFESSION"].str.replace('\n' ,'', regex=True)
    df["PROFESSION"] = df["PROFESSION"].str.replace('Ç' ,'C', regex=True)

    
    df["DATE_ET_LIEU_NAISSANCE"] = df["DATE_ET_LIEU_NAISSANCE"].str.replace('\t' ,'', regex=True)
    df["DATE_ET_LIEU_NAISSANCE"] = df["DATE_ET_LIEU_NAISSANCE"].str.replace('\n' ,'', regex=True)
    df["DATE_ET_LIEU_NAISSANCE"] = df["DATE_ET_LIEU_NAISSANCE"].str.replace('À' ,'A', regex=True)


    df["CONTACTS"] = df["CONTACTS"].str.replace('\t' ,'', regex=True)
    df["CONTACTS"] = df["CONTACTS"].str.replace('\n' ,'', regex=True)

    df["PIECE_IDENTIFICATION"] = df["PIECE_IDENTIFICATION"].str.replace('\t' ,'', regex=True)
    df["PIECE_IDENTIFICATION"] = df["PIECE_IDENTIFICATION"].str.replace('\n' ,'', regex=True)

    df["NUMERO_PIECE"] = df["NUMERO_PIECE"].str.replace('\t' ,'', regex=True)
    df["NUMERO_PIECE"] = df["NUMERO_PIECE"].str.replace('\n' ,'', regex=True)
    df["NUMERO_PIECE"] = df["NUMERO_PIECE"].str.replace(':' ,'', regex=True)
    

    path_csv = path.split(".")[0] + ".csv"
    
    df.to_csv(path_csv, index = False, sep = ";")
    
    return df

    
```
